```python
import pandas as pd
import numpy as np

file_path = "res/vgsales.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df = df.dropna(subset = ['Year'])
df.head()
```

```python
# RELLENANDO PUBLISHER DESCONOCIDO POR EL VALOR DE "UNKNOWN"
df.fillna({'Publisher': 'Unknown'}, inplace = True)

# CAMBIANDO EL TIPO FLOAT DE YEAR A INT
df['Year'] = df['Year'].astype(int)

# SE CALCULA LAS FILAS MAYOR A LA MEDIANA TOTAL DE Global_Sales PARA VER SI x JUEGO ES UN EXITO O NO
df['Market_Trend'] = (df['Global_Sales'] > df['Global_Sales'].median())
df.head(10000)
```
